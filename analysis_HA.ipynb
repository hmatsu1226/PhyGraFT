```r
source("PhyGraFT.R")
```

```r
m <- 100

A <- as.matrix(read.table("dataset_HA/A.txt"))
X <- as.matrix(read.table("dataset_HA/X.txt",header = TRUE))
for(i in 1:dim(X)[2]){
  X[,i] <- X[,i] / sqrt(sum(X[,i]**2))
}

#construct graph Laplacian
L <- sym_normalized_graph_laplacian(A)

#perform GFT
tmp <- graph_fourier_transform(L,X,m)
U <- tmp[[1]]
eigenvalues <- tmp[[2]]
hF <- tmp[[3]]
tF <- tmp[[4]]
```

```r
#plot GF coefficients
barplot_gfdomain(hF,tF,2)
barplot_gfdomain(hF,tF,3)
barplot_gfdomain(hF,tF,4)
```

```r
#plot GLR
plot_aGLR(X, U, 100)
```
